```python
import requests
from bs4 import BeautifulSoup

# כתובת הדף של נווה שרת
url = "https://www.ad.co.il/nadlanrent?sp275=17413&keyword=%D7%A0%D7%95%D7%94+%D7%A9%D7%A8%D7%AA"

# שולחים בקשה עם headers של דפדפן כדי לא להיחסם
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# מחפשים כותרות של דירות
ad_links = []
for a_tag in soup.find_all('a', href=True):
    h2_tag = a_tag.find('h2', class_='card-title mb-0 mb-sm-1')
    if h2_tag and ("נווה שרת" in h2_tag.text or "נוה שרת" in h2_tag.text):
        full_link = "https://www.ad.co.il" + a_tag['href']
        ad_links.append(full_link)

# מדפיסים את הקישורים שמצאנו
for link in ad_links:
    print(link)
```

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np

# מיפוי עברית -> אנגלית עבור שדות מתוך טבלת פרטי הדירה
field_mapping = {
    "פרטי הנכס": "property_type",
    "שכונה": "neighborhood",
    "כתובת": "address",
    "חדרים": "room_num",
    "קומה": "floor",
    "שטח בנוי": "area",
    "שטח גינה": "garden_area",
    "תאריך כניסה": "days_to_enter",
    "תשלומים בשנה": "num_of_payments",
    "ארנונה בחודש": "monthly_arnona",
    "ועד בית בחודש": "building_tax"
}

# תכונות בוליאניות מתוך אייקונים של הדירה
boolean_features = {
    "חניה": "has_parking",
    "מחסן": "has_stotsge",
    "מעלית": "elevator",
    "מזגן": "ac",
    "נגישות": "handicap",
    "סורגים": "has_bars",
    "ממ\"ד": "has_safe_room",
    "מרפסת": "has_balcon",
    "מרוהטת": "is_furnished",
    "משופצת": "is_renovated"
}

all_data = []
headers = {"User-Agent": "Mozilla/5.0"}

# הכנס כאן את ה-API KEY שלך מממשק Google Maps Distance Matrix
API_KEY = "API KEY"

# כתובת היעד לחישוב מרחק – כיכר דיזנגוף בתל אביב
destination_address = "Dizengoff Square, Tel Aviv, Israel"

# ad_links = [...]  ← כאן תכניסי את רשימת הקישורים שלך

for url in ad_links:
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        table = soup.find("table", class_="table table-sm mb-4")
        if not table:
            continue

        flat_data = {}
        found_fields = set()
        days_to_enter_found = False

        for tr in table.find_all('tr'):
            tds = tr.find_all('td')
            if len(tds) != 2:
                continue

            key = tds[0].get_text(strip=True)
            value = tds[1].get_text(strip=True)

            if key in field_mapping:
                eng_key = field_mapping[key]
                found_fields.add(eng_key)

                if eng_key == "floor":
                    if value.strip() == "קרקע":
                        flat_data["floor"] = "0"
                    elif "מתוך" in value:
                        try:
                            floor_val = value.split("מתוך")[0].strip()
                            total_val = value.split("מתוך")[1].strip()
                            flat_data["floor"] = "0" if floor_val == "קרקע" else floor_val
                            flat_data["total_floors"] = total_val
                            found_fields.add("total_floors")
                        except:
                            flat_data["floor"] = value
                    else:
                        flat_data["floor"] = "0" if value.strip() == "קרקע" else value

                elif eng_key == "days_to_enter":
                    days_to_enter_found = True
                    if "מייד" in value:
                        flat_data[eng_key] = 0
                    else:
                        flat_data[eng_key] = np.nan  # אם לא "מיידית", נכניס NaN (אפשר לשנות בהמשך)
                else:
                    flat_data[eng_key] = value

        if not days_to_enter_found:
            flat_data["days_to_enter"] = np.nan

        if "garden_area" not in found_fields:
            flat_data["garden_area"] = "0"
        if "monthly_arnona" not in found_fields:
            flat_data["monthly_arnona"] = "0"

        try:
            desc_tag = soup.find("p", class_="text-word-break")
            flat_data["description"] = desc_tag.get_text(strip=True) if desc_tag else ""
        except:
            flat_data["description"] = ""

        boolean_divs = soup.find_all("div", class_="card-icon col-6 d-inline")
        for div in boolean_divs:
            try:
                icon = div.find("i")
                label_span = div.find("span")
                if not icon or not label_span:
                    continue

                label = label_span.get_text(strip=True)
                if label in boolean_features:
                    field_name = boolean_features[label]
                    icon_class = icon.get("class", [])
                    flat_data[field_name] = 1 if "fa-check" in icon_class else 0 if "fa-times" in icon_class else None
            except:
                continue

        try:
            h2_tags = soup.find_all("h2", class_="card-title")
            if len(h2_tags) >= 2:
                price_text = h2_tags[1].get_text(strip=True)
                price_clean = price_text.replace("₪", "").replace(",", "").strip()
                flat_data["price"] = float(price_clean)
            else:
                flat_data["price"] = 0.0
        except:
            flat_data["price"] = 0.0

        try:
            image_tags = soup.find_all("img", class_="mobile-product-slide")
            flat_data["num_of_images"] = len(image_tags)
        except:
            flat_data["num_of_images"] = 0

        address = flat_data.get("address", None)
        if address:
            full_address = f"{address}, תל אביב, ישראל"
            try:
                distance_url = f"https://maps.googleapis.com/maps/api/distancematrix/json"
                params = {
                    "origins": full_address,
                    "destinations": destination_address,
                    "key": API_KEY,
                    "units": "metric"
                }
                distance_res = requests.get(distance_url, params=params)
                distance_data = distance_res.json()

                if (distance_data["rows"] and
                    distance_data["rows"][0]["elements"] and
                    distance_data["rows"][0]["elements"][0]["status"] == "OK"):

                    distance_km = distance_data["rows"][0]["elements"][0]["distance"]["value"] / 1000
                    flat_data["distance_from_center"] = round(distance_km, 2)
                else:
                    flat_data["distance_from_center"] = np.nan
            except Exception as e:
                print(f"שגיאה בחישוב מרחק עבור {full_address}: {e}")
                flat_data["distance_from_center"] = np.nan
        else:
            flat_data["distance_from_center"] = np.nan

        all_data.append(flat_data)
        time.sleep(1)

    except Exception as e:
        print(f"שגיאה בקישור {url}: {e}")
        continue

# יצירת DataFrame
df = pd.DataFrame(all_data)

# עמודות בוליאניות
bool_columns = [
    "has_parking", "has_stotsge", "elevator", "ac", "handicap",
    "has_bars", "has_safe_room", "has_balcon", "is_furnished", "is_renovated"
]

for col in bool_columns:
    if col in df.columns:
        df[col] = df[col].fillna(0).astype(int)

# עמודות מספריות (כוללת days_to_enter כ־Int64)
int_columns = [
    "floor", "area", "garden_area", "num_of_payments",
    "monthly_arnona", "building_tax", "total_floors", "days_to_enter"
]

float_columns = ["room_num", "price"]

for col in int_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype("Int64")

for col in float_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# סידור העמודות
column_order = [
    "property_type", "neighborhood", "address", "room_num", "floor", "area",
    "garden_area", "days_to_enter", "num_of_payments", "monthly_arnona", "building_tax",
    "total_floors", "description", "has_parking", "has_stotsge", "elevator", "ac",
    "handicap", "has_bars", "has_safe_room", "has_balcon", "is_furnished", "is_renovated",
    "price", "num_of_images", "distance_from_center"
]

existing_columns = [col for col in column_order if col in df.columns]
df = df.reindex(columns=existing_columns)

# המרת object ל-string (למחרוזות בלבד)
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].astype("string")

# שמירה ל־CSV
df.to_csv("apartments.csv", index=False, encoding="utf-8-sig")
# הצגת טבלה
df
```
